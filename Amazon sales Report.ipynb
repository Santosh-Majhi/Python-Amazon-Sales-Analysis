```python
ax=sns.countplot(x="Size", data=df,color="violet")
plt.show()
```

```python
ax=sns.countplot(x="Size", data=df, color="violet")
ax.bar_label(ax.containers[0])
plt.title("count of customers by Size")
plt.show()
```

#from the above Graph, most of the people buys m-size that is 6892.

```python
df.groupby(["Size"], as_index=False)["Qty"].sum().sort_values(by="Qty",ascending= False)
```

```python
s_qty=df.groupby(["Size"], as_index=False)["Qty"].sum().sort_values(by="Qty",ascending= False)
sns.barplot(x="Size", y="Qty",data=s_qty, color="red")
plt.show()
```

#from the above graph, the most of the quantity buys M-size in the sales.

```python
plt.figure(figsize=(7,5))
ax=sns.countplot(x="Courier Status", data=df, hue="Status")
plt.title("Courier Status")
plt.show()
```

#from the above Graph the majority of the orders are shipped through the courier.

```python
df["Size"].hist()
```

```python
df["Category"]=df["Category"].astype(str)
column_data=df["Category"]
plt.figure(figsize=(7,5))
plt.hist(column_data, bins=30, edgecolor="black",color="purple")
plt.xticks(rotation=90)
plt.show()
```

#In this graph the most of the buyers are from T-shirt category.

```python
B2B_check=df["B2B"].value_counts()
plt.pie(B2B_check, labels=B2B_check.index, autopct="%1.1f%%")
plt.show()
```

#from the above chart the maximum buyers i.e 99.2% are retailers and 0.8% are B2B buyers.

```python
x_data=df["Category"]
y_data=df["Size"]
plt.scatter(x_data, y_data, color="gold")
plt.xlabel("Category")
plt.ylabel("Size")
plt.title("Scatter Plot on the Basis of Category and Size")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df, x="ship-state", color="lime")
plt.xlabel("Ship-State")
plt.ylabel("Count")
plt.title("Distribution of State")
plt.xticks(rotation=90)
plt.show()
```

```python
top_10_state=df["ship-state"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.countplot(data=df[df["ship-state"].isin(top_10_state.index)], x="ship-state", color="magenta")
plt.xlabel("Ship-State")
plt.ylabel("Count")
plt.title("Distribution of State")
plt.xticks(rotation=45)
plt.show()
```

#from the above graph, the most of the buyers are from maharastra state.
